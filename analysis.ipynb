```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df_issues = pd.read_csv('refactoring_analysis_general.csv')
df_issues_detail = pd.read_csv('refactoring_analysis_details.csv')
df_git = pd.read_csv('codesamples.csv')
```

```python
df_issues_detail['owner'] = df_issues_detail['repository'].apply(lambda x: df_git[df_git['html_url'] == x]['owner'].item() if len(df_git[df_git['html_url'] == x]) else '')
```

```python
df_git_aux = df_git[df_git['html_url'].isin(df_issues_detail['repository'])].set_index('html_url')#[['total_lines', 'first_commit_date', 'last_commit_date', 'contributors_count']]
n_df_issues_detail = df_issues_detail.set_index('repository')#.groupby('repository').count()[['sha1']]
n_df_issues_detail = n_df_issues_detail.rename(columns={'key': 'n_df_issues_detail'})
n_df_issues_detail = n_df_issues_detail.join(df_git_aux, lsuffix="1")
n_df_issues_detail['first_commit_date'] = pd.to_datetime(n_df_issues_detail['first_commit_date'])
n_df_issues_detail['last_commit_date'] = pd.to_datetime(n_df_issues_detail['last_commit_date'])
n_df_issues_detail['total_time'] = (n_df_issues_detail['last_commit_date'] - n_df_issues_detail['first_commit_date']).dt.days
n_df_issues_detail.tail()
```

```python
n_df_issues_detail['stargazers_count']
```

```python
aux = n_df_issues_detail[n_df_issues_detail['commits_count'] != 0]
media_line = aux['sha1']/(aux['commits_count']/1)
media_line.hist(bins=60)
plt.plot(n_df_issues_detail['commits_count'], n_df_issues_detail['sha1'], 'o', label='Número de Code Smells Encontrados')  

plt.legend()
plt.show()
media_line.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
cleaned_data = n_df_issues_detail.dropna(subset=['total_time', 'sha1'])
cleaned_data = cleaned_data[cleaned_data['sha1'] < 350]
# Dados de exemplo
x = cleaned_data['total_time'].values.reshape(-1, 1)  # Reshape necessário para regressão
y = cleaned_data['sha1'].values
print(y.max())

# Criando e treinando o modelo de regressão linear
reg = LinearRegression()
reg.fit(x, y)

# Gerando valores previstos para a linha de regressão
x_pred = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)  # Valores para plotar a linha
y_pred = reg.predict(x_pred)

# Plotando os pontos e a linha de regressão
plt.figure(figsize=(8, 6))
plt.plot(cleaned_data['total_time'], cleaned_data['sha1'], 'o', label='Número de Refatorações Realizadas', color='orange')
plt.plot(x_pred, y_pred, '-', label='Regressão Linear', color='blue')

# Configurando o gráfico
plt.xlabel("Tempo de Projeto")
plt.ylabel("Número de Refatorações Realizadas")
plt.title("Regressão Linear: Tempo de Projeto vs Refatoração")
plt.legend()
plt.grid(True)
plt.show()

# Coeficientes e intercepto
print(f"Coeficiente angular (slope): {reg.coef_[0]:.2f}")
print(f"Intercepto (intercept): {reg.intercept_:.2f}")
```
